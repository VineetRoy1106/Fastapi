##**Task 1**

```python
# Filtering for pageview events and calculating total and average
total_pageviews = df_cleaned[df_cleaned['event'] == 'pageview'].shape[0]
total_pageviews
```

```python
daily_pageviews = df_cleaned[df_cleaned['event'] == 'pageview'].groupby('date').size()
daily_pageviews
```

##**Task 4**

```python
# Calculating CTR (clicks/pageviews)
total_clicks = df_cleaned[df_cleaned['event'] == 'click'].shape[0]
ctr = total_clicks/total_pageviews
ctr
```

```python
type(total_pageviews)
```

```python
# CTR variation by link
ctr_by_link = df_cleaned[df_cleaned['event'].isin(['click', 'pageview'])].groupby(['linkid', 'event']).size().unstack(fill_value=0)
ctr_by_link['CTR'] = ctr_by_link['click'] / ctr_by_link['pageview']
ctr_by_link
```

##**Task 5**

```python
# Creating a contingency table for categorical correlation analysis
contingency_table = df_cleaned.pivot_table(index='linkid', columns='event', aggfunc='size', fill_value=0)
contingency_table['correlation_coefficient'] = contingency_table['click'].corr(contingency_table['pageview'])
contingency_table
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Prepare data for analysis (using 'linkid' to aggregate clicks and previews)
clicks_previews = df_cleaned[df_cleaned['event'].isin(['click', 'pageview'])].pivot_table(
    index='linkid', columns='event', aggfunc='size', fill_value=0
)

# Calculate Pearson correlation coefficient
correlation, p_value = pearsonr(clicks_previews['click'], clicks_previews['pageview'])

# Linear regression
X = clicks_previews['pageview'].values.reshape(-1, 1)  # Independent variable (previews)
y = clicks_previews['click'].values  # Dependent variable (clicks)

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# R-squared value
r_squared = r2_score(y, y_pred)

# Visualizing the linear relationship
plt.figure(figsize=(10, 6))
sns.scatterplot(x='pageview', y='click', data=clicks_previews, alpha=0.6)
plt.plot(clicks_previews['pageview'], y_pred, color='red', label=f"Linear Fit (R²={r_squared:.2f})")
plt.title('Linear Relationship between Previews and Clicks')
plt.xlabel('Previews (Pageviews)')
plt.ylabel('Clicks')
plt.legend()
plt.grid()
plt.show()

{
    "Pearson Correlation": correlation,
    "P-Value": p_value,
    "R-Squared": r_squared
}
```

Pearson Correlation:

1.   Correlation Coefficient: 0.994
      
      This indicates a strong positive linear relationship between previews (pageviews) and clicks.


2.  P-Value: 0.0

    The result is statistically significant, confirming the correlation is unlikely due to chance.



3.   R-Squared: 0.988

    Approximately 98.8% of the variability in clicks is explained by the
    linear relationship with previews.

Interpretation:
  
  

*   The analysis strongly supports that previews (pageviews) and clicks   are linearly related.
*   A well-fitted linear regression model can reliably predict clicks from previews.
